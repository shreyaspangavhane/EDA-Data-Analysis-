```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.columns
```

```python
df.columns
```

```python
# combining of 2 dataframe on basic of country code and use left inner join
    # merge parameter  (leftsidedata,rightsidedata,on=common_part,how=joins name)

final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes              # check the data types
```

```python
# Que-> check in which country zomato mostly used
final_df.Country.value_counts()         #tell number of records
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_names_count=final_df.Country.value_counts().values
```

```python
# sns.histplot(x="country_names",y="country_names_count",data=final_df)

#pie chart ->top 3 countries that uses zomato
plt.pie(country_names_count[:3],labels=country_names[:3],autopct="%.2f%%")
plt.show()
```

Observation: Zomato mostly used by Indian Peopel as comapre to USA then UK

```python
final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index()      #make indexing
```

```python
ratings=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

## Observation:
1. When rating is between 4.5 to 4.9---> Excellent
2. When rating is between 4.0 to 4.4---> Very good
3. When rating is between 3.5 to 3.9---> good
4. When rating is between 2.5 to 3.4---> Average
4. When rating is between 1.8 to 2.4---> Poor

```python
ratings.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings)
```

    Observation: noit in color as it shiws in that box of scale

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','darkgreen'])
```

## Observation:
      1. Not rated count is very high 
      2. Maximum rating is between 2.5 to 4.0

```python
##Count plot   -> count no of each color of rating
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','darkgreen'])
```
